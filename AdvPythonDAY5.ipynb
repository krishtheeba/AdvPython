```python
house_price=np.array([1000,2000,3000,4000,500])
sqft=np.array([500,1000,1500,2000,250])

import matplotlib.pyplot as plt

plt.scatter(house_price,sqft,marker="+")
plt.xlabel("HOUSE PRICE")
plt.ylabel("SQUARE FEET")   # EDA - Exploratory Data Analysis - bivarient analysis
plt.title("HOUSE PRICE PREDICTION ")
```

```python
n1=np.array([1,2,3,4,5])
n2=np.array([2,4,6,7,6])
plt.bar(n1,n2)   # bar 
plt.xlabel("Hours of Study")
plt.ylabel("Marks Obtained")
plt.title("Mark Analysis with Study Hours")
```
