```python
import threading
import requests
import bs4
import re
def f1(url):
    print("URL:{}".format(url))
    for url in url:
        r=requests.get(url)
        if r.status_code != 200:
            print("Download Failed")
            exit()
        page=r.text
        f2(page)
def f2(page):
    obj=bs4.BeautifulSoup(page)
    for var in obj.find_all('a'):
        if re.search("^(http|https)",var.get('href')):
            with open("C:\\Users\\Theeba\\Desktop\\urls.log","a") as WH:
                if var.get('href') is str:
                    WH.write(str(var.get('href'))+"\n")

tobj=threading.Thread(target=f1,args=(('http://www.python.org', 'http://www.google.com' , 'http://www.oracle.com'),))
tobj.start()
tobj.join()
print("Exit from Main Script")
```

```python
import threading
import requests
import bs4
import re
import time
def f1(url):
    print("URL:{}".format(url))
    time.sleep(1)
    for url in url:
        r=requests.get(url)
        if r.status_code != 200:
            print("Download Failed")
            exit()
        page=r.text
        f2(page)
def f2(page):
    obj=bs4.BeautifulSoup(page)
    for var in obj.find_all('a'):
        if re.search("^(http|https)",var.get('href')):
            with open("C:\\Users\\Theeba\\Desktop\\urls.log","a") as WH:
                if var.get('href') is str:
                    WH.write(str(var.get('href'))+"\n")

Urls=('http://www.python.org', 'http://www.google.com' , 'http://www.oracle.com')
for v in Urls:
    tobj=threading.Thread(target=f1,args=(v,))
    tobj.start()
    tobj.join()
print("Exit from Main Script")
```

```python
import bs4
import re
import time
def f1(url):
    print("URL:{}".format(url))
    time.sleep(1)
    for url in url:
        r=requests.get(url)
        if r.status_code != 200:
            print("Download Failed")
            exit()
        page=r.text
        f2(page)
def f2(page):
    obj=bs4.BeautifulSoup(page)
    txt=obj.get_text()
    print(txt)
    time.sleep(1)

Urls=('http://www.python.org', 'http://www.google.com' , 'http://www.oracle.com')
for v in Urls:
    tobj=threading.Thread(target=f1,args=(v,))
    tobj.start()
    tobj.join()
print("Exit from Main Script")
```
